```python
def compute_adjusted_weight(edge, a=0.8, b=1.2):
    highway_type = edge[2]['highway']
    length = float(edge[2]['length'])
    
    if highway_type in ['primary', 'primary_link', 'secondary']:
        return length * a  
    else:
        return length * b  

for edge in subiaco_graph.edges(data=True):
    edge[2]['adjusted_weight'] = compute_adjusted_weight(edge)
```

```python
def find_path(G, source, target, weight='adjusted_weight'):
    return nx.shortest_path(G, source=source, target=target, weight=weight)

def check_time_patterns(time_step):
        if (time_step >= 200) and (time_step < 300):
            time_pattern = "morning_peak"
        elif (time_step >= 700) and (time_step < 800):
            time_pattern = "evening_peak"
        else:
            time_pattern = "normal"       
        return time_pattern
    
    
def choose_node(reverse=True):
    nodes = [level_1, level_2, level_3, level_4]
    
    if reverse:
        weights = [0.4, 0.4, 0.1, 0.1] # morning peak des & evening peak star
    else:
        weights = [0.1, 0.1, 0.4, 0.4] # morning peak star & evening peak des
    
    return random.choices(nodes, weights, k=1)[0]


def initialize_start_and_dest(pattern):
    if pattern == 'morning_peak':
        start = random.sample(choose_node(reverse = False), 1)[0]
        destination = random.sample(choose_node(reverse = True), 1)[0]
    elif pattern == 'evening_peak':
        start = random.sample(choose_node(reverse = True), 1)[0]
        destination = random.sample(choose_node(reverse = False), 1)[0]
    else:
        start, destination = random.sample(all_nodes, 2)
    return start, destination 

          
def check_all_vehicles_arrived(vehicles):
    arrived = True
    for vehicle in vehicles:
        if vehicle.status != "arrived":
            arrived = False
            return arrived
    return arrived


def get_road_speed_limit(graph, current_position, next_position):
    """Get the speed limit of the road on which the vehicle is currently positioned."""
    maxspeed = graph.get_edge_data(current_position, next_position)[0]['maxspeed']
    if maxspeed:
        return maxspeed   
    return 60.0  # default speed limit


def edges_connected_to_node(graph, node):
     return list(graph.edges(node))
    
    
def check_provided_nodes(graph, nodes:list):
    for node in nodes:
        if len(list(graph.edges(node))) <=2:
            return 0
    return 1


def check_wether_pass(next_position, traffic_light, time_step):
    edge, stay_time = traffic_light.get_current_duration(time_step)
    if edge == next_position:
        return True
    else:
        return False

    
def divide_into_levels(node_list):
    level1 = []
    level2 = []
    level3 = []
    level4 = []
    for i in node_list:
        if i in level_1:
            level1.append(pos[i])
        elif i in level_2:
            level2.append(pos[i])
        elif i in level_3:
            level3.append(pos[i])
        else:
            level4.append(pos[i])
    return [level1, level2, level3, level4]


def make_circular(vehicle_dest_level):
    level_circular = []
    for level in vehicle_dest_level:
        avg_latitude = sum(lat for lat, _ in level) / len(level)
        avg_longitude = sum(lon for _, lon in level) / len(level)
        vehicles_num = len(level)
        level_circular.append([[avg_latitude, avg_longitude], vehicles_num])  
    return level_circular

    
def send_vehicle_data_to_server(graph):
    # Extract necessary data from the Vehicle objects
    vehicles = graph.vehicles
    
    vehicle_current_position = [list(graph.pos[vehicle.current_position]) for vehicle in graph.vehicles]
    vehicle_speed_ratio = [vehicle.speed_ratio for vehicle in graph.vehicles]
    vehicle_dest_level = divide_into_levels([str(vehicle.destination) for vehicle in graph.vehicles])
    
    level_circular = make_circular(vehicle_dest_level)
    vehicle_data = [vehicle_current_position, vehicle_speed_ratio]
    if len(vehicle_current_position) != len(vehicle_speed_ratio):
        print("data is wrong!")
    
    url_vehicles = "http://127.0.0.1:5000/update_positions"
    url_circle = "http://127.0.0.1:5000/update_circle_positions"
    try:
        response = requests.post(url_vehicles, json=vehicle_data)
        if response.status_code != 200:
            print(f"Failed to send data. Server responded with: {response.status_code}")
    except requests.ConnectionError:
        print("Failed to connect to the server. Make sure the server is running.")
        
    try:
        response = requests.post(url_circle, json=level_circular)
        if response.status_code != 200:
            print(f"Failed to send data. Server responded with: {response.status_code}")
    except requests.ConnectionError:
        print("Failed to connect to the server. Make sure the server is running.")
        
```

```python
class traffic_flow_map():
    def __init__(self, graph, normal_vehicles_num, traffic_light_num):
        self.graph = graph
        self.time_step = 0
        self.vehicles = []
        self.traffic_light_list = []
        self.normal_vehicles_num = normal_vehicles_num
        
        self.traffic_flow = {}
        self.vehicle_positions = {}
        self.congestion_vehicles = {}
        self.congestion_nodes = {}
        
        self.pos = {node: (data['x'], data['y']) for node, data in graph.nodes(data=True)}
        
        # Initialize vehicles and traffic lights only once in the constructor
        self.initialize_vehicles(normal_vehicles_num)
        self.initialize_traffic_light(traffic_light_num = traffic_light_num)
            
            
    def initialize_vehicles(self, vehicles_num):
        self.vehicles = [Vehicle(self.graph, self.traffic_flow, self.time_step) for _ in range(vehicles_num)]
            
        
    def initialize_traffic_light(self, traffic_light_num, nodes=None):
        traffic_light_node_list = []
        
        if nodes and check_provided_nodes(self.graph, nodes):
            traffic_light_node_list = [(node, self.graph.nodes[node]) for node in nodes] 
        else:
            candidate_nodes = [node for node in self.graph.nodes() if len(list(set(self.graph.edges(node)))) > 2]
            traffic_light_node_list = random.sample(candidate_nodes, min(traffic_light_num, len(candidate_nodes)))
            
        self.traffic_light_list = [Traffic_light(self.graph, node, self.time_step) for node in traffic_light_node_list]
    
    
    def record_traffic_flow(self):
        current_flow = {}
        for i in self.vehicles:
            key = (i.current_position, i.next_position)
            if key in current_flow:
                current_flow[key] += 1
            else:
                current_flow[key] = 1
                
        for key, value in current_flow.items():
            if key not in self.traffic_flow:
                self.traffic_flow[key] = {}
                self.traffic_flow[key][self.time_step] = value
            else:
                self.traffic_flow[key][self.time_step] = value
 

    def record_vehicles_position(self):
        self.vehicle_positions[self.time_step] = []
        for vehicle in self.vehicles:
            if vehicle.current_position in pos.keys():
                self.vehicle_positions[self.time_step].append([vehicle.current_position, pos[vehicle.current_position][1], pos[vehicle.current_position][0]])
            
            if vehicle.speed_ratio < 0.2:
                if self.time_step in self.congestion_vehicles.keys():
                    self.congestion_vehicles[self.time_step].append([vehicle.current_position, vehicle.next_position, vehicle.speed_ratio, self.traffic_flow[vehicle.current_position, vehicle.next_position][self.time_step-1]])
                else:
                    self.congestion_vehicles[self.time_step] = []
                    self.congestion_vehicles[self.time_step].append([vehicle.current_position, vehicle.next_position, vehicle.speed_ratio, self.traffic_flow[vehicle.current_position, vehicle.next_position][self.time_step-1]])
 
    
    def eliminate_congestion(self):
        vehicle_positions = set(tuple(item) for item in graph.vehicle_positions[self.time_step])
        for i in vehicle_positions:
            count = 0 
            for item in graph.vehicle_positions[self.time_step]:
                if i == tuple(item):
                    count += 1
            if  count > 15:
                vehicle_congestion_list = []
                for vehicle in self.vehicles:
                    if vehicle.current_position == i[0] and vehicle.speed_ratio < 0.2:
                        vehicle_congestion_list.append(vehicle)
                for vehicle in vehicle_congestion_list:
                    vehicle.next_time_step = int(vehicle.next_time_step * 0.6) + 1
                if self.time_step in self.congestion_nodes.keys():
                    self.congestion_nodes[self.time_step].append([i[0], count])
                else:
                    self.congestion_nodes[self.time_step] = []
                    self.congestion_nodes[self.time_step].append([i[0], count])
    

    def animate(self, time_step):
        max_vehicles = self.normal_vehicles_num + 5
        min_vehicles = self.normal_vehicles_num - 5
        
        traffic_light = list(map(list, [self.pos[trafflic_light.node] for trafflic_light in self.traffic_light_list]))
        requests.post('http://127.0.0.1:5000/update_traffic_lights', json=traffic_light)
 
        while self.time_step < time_step:  
            # Update each traffic light's status based on the current time step
            for traffic_light_obj in self.traffic_light_list:
                traffic_light_obj.get_current_duration(self.time_step)
        
            # Move each vehicle based on the current traffic lights and time step
            for vehicle in self.vehicles:
                vehicle.move(self.traffic_light_list, self.traffic_flow, self.time_step)
            
            self.vehicles = [vehicle for vehicle in self.vehicles if vehicle.status != "arrived"]
            
            if check_time_patterns(self.time_step) != 'normal':
                max_vehicles = self.normal_vehicles_num*2 + 10
                min_vehicles = self.normal_vehicles_num*2 - 10
            else:
                max_vehicles = self.normal_vehicles_num + 5 
                min_vehicles = self.normal_vehicles_num - 5
            
            if len(self.vehicles) < max_vehicles:
                to_add = random.randint(0, (max_vehicles - len(self.vehicles)))
                
                if len(self.vehicles) + to_add > min_vehicles:
                    to_add = min(5, to_add)
                else:
                    to_add = (min_vehicles - len(self.vehicles))
                    to_add = min(10, to_add)
                    
                for _ in range(to_add):
                    new_vehicle = Vehicle(self.graph, self.traffic_flow, self.time_step)
                    self.vehicles.append(new_vehicle)
                    
            
            
            self.record_traffic_flow()
            self.record_vehicles_position()
            
            if self.time_step % 10 == 0:
                self.eliminate_congestion()
            
            send_vehicle_data_to_server(self)
            self.time_step += 1
            print(self.time_step)
            time.sleep(0.1)
          
        return
```

```python
class Vehicle:
    def __init__(self, graph, traffic_flow, time_step):
        self.graph = graph
        self.status = "moving"  # can be "moving", "arrived", "intersection"
        
        self.initialize_position(time_step)
        
        self.initialize_path()
        
        self.speed = 0  # Initial speed
        self.speed_ratio = 0
        self.next_time_step = 0 # Initial time step
        
        self.assume_next_time_step(traffic_flow, time_step)
               
        
    def initialize_position(self, time_step):
            time_pattern = check_time_patterns(time_step)
            self.start, self.destination = initialize_start_and_dest(time_pattern)
            while (not nx.has_path(self.graph, self.start, self.destination)) or (len(nx.shortest_path(self.graph, source=self.start, target=self.destination, weight='length')) < 2):
                 self.start, self.destination = initialize_start_and_dest(time_pattern)
            self.current_position = self.start   
            
            
    def initialize_path(self):
        #self.path = nx.shortest_path(self.graph, source=self.start, target=self.destination, weight='length')
        self.path = find_path(self.graph, source=self.start, target=self.destination)
        self.next_position = self.path[1]
        
        
    def assume_next_time_step(self, traffic_flow, time_step):
        edge_data = self.graph.get_edge_data(self.current_position, self.next_position)[0]
        length = edge_data.get('length', 100) 
        road_type = edge_data.get('highway', 'unknown')
        if road_type not in ['primary', 'primary_link', 'secondary']:
            vehicle_limited = 20
        else: vehicle_limited = 30
        
        vehicles_on_road = traffic_flow.get((self.current_position, self.next_position), {}).get(time_step-1, 0)
        speed_limited = (vehicle_limited - vehicles_on_road) / vehicle_limited
        speed_limited = 0.1 if speed_limited <= 0 else speed_limited
        
        road_speed_limit = get_road_speed_limit(self.graph, self.current_position, self.next_position)
        maxspeed = road_speed_limit * speed_limited
        
        self.speed = random.uniform(0.8*maxspeed, maxspeed)
        self.speed_ratio = self.speed / road_speed_limit
        self.next_time_step =  math.ceil(length / (self.speed / 3.6)) 
    
    
    def check_traffic_light(self, traffic_light_list, time_step):
        for traffic_light_obj in traffic_light_list:
            if self.current_position == traffic_light_obj.node:
                return check_wether_pass(self.next_position, traffic_light_obj, time_step)
        return True        
    
    
    def move(self, traffic_light_list, traffic_flow, time_step):
        if self.status == "arrived":
            return
       
        if self.status == "intersection":
            if self.check_traffic_light(traffic_light_list, time_step):
                self.status = "moving"
                self.next_time_step -= 1
                
            
        if self.status == "moving":
            if self.next_time_step == 0:
                if self.next_position == self.destination:
                    self.current_position = self.destination
                    self.status = "arrived"
                else:
                    self.current_position = self.next_position
                    self.next_position = self.path[self.path.index(self.current_position) + 1]
                    self.assume_next_time_step(traffic_flow, time_step)
                    self.status = "intersection"
            else:
                self.next_time_step -=1           
    
```

```python
class Traffic_light():
    def __init__(self, graph, node, time_step):
        self.node = node
        self.connected_edges = list(set(graph.edges(self.node)))
        self.time_pattern = None  
        self.total_cycle_time = 0
        
        # Get road types of connected edges
        self.road_types = [graph[edge[0]][edge[1]].get('highway', 'unknown') for edge in self.connected_edges]
        self.check_time_patterns(time_step)
        self.determine_light_durations()   
    

    def check_time_patterns(self, time_step):
        if (time_step >= 200) and (time_step < 300):
            self.time_pattern = "morning_peak"
        elif (time_step >= 700) and (time_step < 800):
            self.time_pattern = "evening_peak"
        else:
            self.time_pattern = "normal"       
        return
        
        
    def determine_light_durations(self):
        # Light durations based on road types
        light_durations = {
            'primary': 10,
            'primary_link': 15,
            'secondary': 10,
            'tertiary': 5, #25
            'residential': 5,
            'unclassified': 5,
            'unknown': 5
        }
        
        
        if self.time_pattern in ["morning_peak", "evening_peak"]:
            self.road_durations = [light_durations.get(road_type, 5) + 5 for road_type in self.road_types]
        else:
            self.road_durations = [light_durations.get(road_type, 5) for road_type in self.road_types]
            
        self.total_cycle_time = sum(self.road_durations)
        self.durations = {edge[1]: duration for edge, duration in zip(self.connected_edges, self.road_durations)}        
        
    
    def get_current_duration(self, time_step):
        self.check_time_patterns(time_step)
        self.determine_light_durations()
        
        accumulated_time = 0
        current_time_in_cycle = time_step % self.total_cycle_time
        
        for edge, duration in self.durations.items():
            accumulated_time += duration
            if current_time_in_cycle < accumulated_time:
                stay_time = accumulated_time - current_time_in_cycle
                return edge, stay_time
```

```python
def nodes_in_grid(grid, nodes):
    """Count the number of nodes inside a grid."""
    count = 0
    for node in nodes:
        lat, lon = float(node[1]), float(node[2])
        if grid[0] <= lat <= grid[1] and grid[2] <= lon <= grid[3]:
            count += 1
    return count

def plot_vehicle_density(grid_boundaries, nodes_by_step, interval):
    # Make sure the path exists
    if not os.path.exists('./density_plot'):
        os.mkdir('./density_plot')
    time_step = 0
    while time_step in nodes_by_step.keys():
        # Count the number of nodes in each grid
        node_counts = [nodes_in_grid(grid, nodes_by_step[time_step]) for grid in grid_boundaries]
        max_count = max(node_counts)
        
        # Clear the previous output
        clear_output(wait=True)
        
        plt.figure(figsize=(12, 8))
        # Coloring the grids based on node counts using the Cold Map scheme
        for boundary, count in zip(grid_boundaries, node_counts):
            color_intensity = count / max_count
            plt.fill(
                [boundary[2], boundary[3], boundary[3], boundary[2], boundary[2]],
                [boundary[0], boundary[0], boundary[1], boundary[1], boundary[0]],
                color=plt.cm.Blues(color_intensity), edgecolor='red', linewidth=0.5
            )

        plt.xlabel('Longitude')
        plt.ylabel('Latitude')
        plt.title(f'Grids based on Vehicle Density (time step:{time_step})')
        plt.grid(True, which='both', linestyle='--', linewidth=0.5)
        plt.tight_layout()
        
        # If time_step is a multiple of 50, save the plot
        if time_step % 50 == 0:
            plt.savefig(f'./density_plot/density_{time_step}.png')
        
        # Display the current plot
        display(plt.gcf())
        time.sleep(1)
        time_step += interval
    return
```

```python
def extract_traffic_data(street_name, file_path):
    # Initialize an empty list to store the extracted data
    street_data = []
    
    # Open and read the CSV file
    with open(file_path, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)  # Using DictReader to read the file into a list of dictionaries
        
        # Extracting the required data for the specified street
        for row in reader:
            if row['STREET_LOCATION'] == street_name:
                street_data.append({
                    'AVE_WEEKDAY_TRAFFIC': float(row['AVE_WEEKDAY_TRAFFIC']),
                    'AM_PEAK_8TO9': float(row['AM_PEAK_8TO9']),
                    'PM_PEAK_5TO6': float(row['PM_PEAK_5TO6'])
                })
    
    return street_data

def calculate_averages(traffic_data):
    # Initialize variables to store the total values
    total_ave_weekday_traffic = 0
    total_am_peak = 0
    total_pm_peak = 0
    
    # Sum up the values
    for data in traffic_data:
        total_ave_weekday_traffic += data['AVE_WEEKDAY_TRAFFIC']
        total_am_peak += data['AM_PEAK_8TO9']
        total_pm_peak += data['PM_PEAK_5TO6']
    
    # Calculate averages
    num_entries = len(traffic_data)
    avg_ave_weekday_traffic = total_ave_weekday_traffic / num_entries
    avg_am_peak = total_am_peak / num_entries
    avg_pm_peak = total_pm_peak / num_entries
    
    return {
        'Average AVE_WEEKDAY_TRAFFIC': avg_ave_weekday_traffic,
        'Average AM_PEAK_8TO9': avg_am_peak,
        'Average PM_PEAK_5TO6': avg_pm_peak
    }

hay_street = extract_traffic_data('Hay Street', './traffic_flow_data.csv')
ave_hay_street_flow = calculate_averages(hay_street)
ave_hay_street_flow
```
